### API PULL


```python
import pandas as pd
import requests
import datetime as dt

##https://docs.helium.com/api/blockchain/hotspots

datapulls = []
hotspotaddys = ['11LzdNQ2Xa8QqXUs6iMVRXLHbcLEdiFAs1TnPyge1CR787ts5DL','11ywGrcVj57dV7SAr1d1JuXu8257fD8cGGtKrJ7QnyrS7eQuSfN']
for i in hotspotaddys:
    URL = "https://api.helium.io/v1/hotspots/" + i + "/activity"
    PARAMS = []
    requestdata = {'data': [], 'cursor': []}
    ticker = 0
    ##globals()[f"{i}_ticker"] = 0
    while requestdata['data'] == [] or len(requestdata) == 2:
        r = requests.get(url = URL,params = PARAMS)
        requestdata = r.json()
        globals()[f"requestdata_{i}_{ticker}"] = requestdata
        ticker += 1
        if len(requestdata)== 2:
            PARAMS = {'cursor': requestdata['cursor']}
        else:
            PARAM = []
        print('request' + str(ticker) + ' for ' + str(i))
        ##datapulls.append(str(globals()[f"requestdata_{i}{ticker}"]))
    print('success')
```

### Create Dataframe from Json Files

```python
datapulls = list(filter(lambda k: 'requestdata_' in k, dir()))
masterdata = pd.DataFrame(columns = ['type_x', 'gateway', 'amount', 'account', 'joinkey', 'type_y', 'time','start_epoch', 'rewards', 'height', 'hash', 'end_epoch', 'secret','request_block_hash', 'path', 'onion_key_hash', 'fee','challenger_owner', 'challenger_lon', 'challenger_location','challenger_lat', 'challenger', 'state_channel', 'conflicts_with','closer', 'version', 'secret_hash', 'block_hash'])

for d in range(0,len(datapulls)):
    if len(datapulls[d]) == 0:
        bn = pd.DataFrame.from_records(datapulls[d]['data'])
        bn = bn.dropna(axis=0, subset=['rewards'])
        bn = bn.reset_index(drop=True)
        bn['joinkey'] = bn.index

        tempmaster = pd.DataFrame(columns = ['type', 'gateway','amount','account','joinkey'])
        for i in range(0,len(bn.index)):
            df = pd.DataFrame(bn['rewards'][i])
            df['joinkey'] = i
            tempmaster = pd.concat([tempmaster, df])

        tempmaster = pd.merge(tempmaster,bn,how= 'left', on=['joinkey'])
        tempmaster.drop(['joinkey'], axis=1)
        tempmaster = tempmaster.reset_index(drop=True)
    masterdata = pd.concat([masterdata, tempmaster])

masterdata.head()
```

### Clean up Rewards Table

```python
##table insert timestamp
masterdata['insert_timestamp'] = dt.datetime.now()
##reset index
masterdata = masterdata.reset_index(drop=True)
##drop columns that dont have reward values
masterdata = masterdata.drop(['joinkey','secret','request_block_hash', 'path', 'onion_key_hash', 'fee','challenger_owner', 'challenger_lon', 'challenger_location','challenger_lat', 'challenger', 'state_channel', 'conflicts_with','closer', 'version', 'secret_hash', 'block_hash'], axis=1)
##rename columns for clarity
masterdata = masterdata.rename(columns={"type_x": "reward_subtype", "type_y": "reward_type","time":"reward_timestamp","rewards":"reward_json","height":"block_height"})
##reorder
masterdata = masterdata[['insert_timestamp','hash','block_height','start_epoch','end_epoch','gateway','account','reward_timestamp','reward_type','reward_subtype','amount','reward_json']]

masterdata.head()
```
